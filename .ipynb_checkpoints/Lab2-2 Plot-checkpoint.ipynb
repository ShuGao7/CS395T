```python
# You can also fill in
#import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

plt.show()
```

```python
# Fillin between y1 & y2
#import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

```python
# Scatter Plot
#import matplotlib.pyplot as plt
#import numpy as np

np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
# 3Dimesion

n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

xs = (xmax - xmin) * np.random.rand(n) + xmin
ys = (xmax - xmin) * np.random.rand(n) + ymin
zs = (xmax - xmin) * np.random.rand(n) + zmin
color = (xmax - xmin) * np.random.rand(n) + cmin

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')
```

```python
#heatmap

import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))
# cmap = plt.get_cmap('PiYG')
# cmap = plt.get_cmap('BuGn')
# cmap = plt.get_cmap('Greys')
cmap = plt.get_cmap('bwr')

plt.matshow(arr, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
#Box Plot

# 1. Set Style
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. Data
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. Graph
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
#I want to foucs only the age and fare
titanic_df2 = titanic_df.loc[:,['Age','Fare']]
```

```python
titanic
```

```python
x = titanic[:,[0]]
y = titanic[:,[1]]
```

```python
#Look at the arrays. Printing to the notebook output works for small datasets, but is not practical for large datasets
print(x)
print('\n')
print(y)
```

```python
plt.plot(x,y,'k') #'k' indicates the line is the color black.
```

```python
#Create a figure with 2 subplots which are stored in axs
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (12,6))

#Notice that we can manipulate the color and marker by passing arguments to scatter

#First subplot is modified by calling methods on axs[0]
axs[0].scatter(x, y, marker = '*', color = 'k')
axs[0].set_title('Scatter Plot Black Stars')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

#Second subplot is modified by calling methods on axs[1]
axs[1].scatter(x, y, marker = '^', color = 'r')
axs[1].set_title('Scatter Plot Red Triangles')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
```
